```python
def Polynomial():
    print("Welcome to the polynomial calculator.")
    # Accept user input as Polynomial
    # Note: Please enter the valuse as 3*x**2 instead of 3x^2

    input1 = sym.Poly(input("Please enter the first Polynomial: "))
    input2 = sym.Poly(input("Please enter the second Polynomial: "))
    p1= input1
    p2 = input2

    # Accept operation type
    user_input = input("Please, choose Operation (+, -, *, /): ")
    # Print the result
    if user_input == '+':
        print("Result = ")
        display(Math('(%s) + (%s) =%s'%(sym.latex(p1),sym.latex(p2),sym.latex(p1 + p2))))
    elif user_input == '-':
        print("Result = ")
        display(Math('(%s) - (%s) =%s'%(sym.latex(p1),sym.latex(p2),sym.latex(p1 - p2))))
    elif user_input == '*':
        print("Result = ")
        display(Math('(%s) * (%s) =%s'%(sym.latex(p1),sym.latex(p2),sym.latex(p1 * p2))))
    elif user_input == '/':
        print("Result = ")
        display(Math('(%s) / (%s) =%s'%(sym.latex(p1),sym.latex(p2),sym.latex(p1 / p2))))
    else:
        print("Unsuppoerted operator. Restart the code! ")
        Polynomial()
    

```

```python
def Repeat():
    repeat = input("Choose: n to enter new arguments, x to exit: ")
    if repeat == 'n'.lower():
        return Polynomial()
    elif repeat  == 'x':
        print("Bye!")
        

    
```

```python
while True:
    Polynomial()
    Repeat()
    break
    
    
    
```
